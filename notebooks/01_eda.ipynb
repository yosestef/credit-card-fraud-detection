```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Cargar el dataset
df = pd.read_csv("../data/creditcard.csv")

# Gráfica de barras: cuántas transacciones de cada clase
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Class")
plt.title("Transacciones legítimas vs fraudulentas")
plt.xlabel("Clase (0 = legítima, 1 = fraude)")
plt.ylabel("Cantidad de transacciones")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df["Amount"], bins=50, kde=True)
plt.title("Distribución del monto de las transacciones")
plt.xlabel("Monto ($)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=df, x="Class", y="Amount")
plt.title("Comparación de montos: legítimas vs fraude")
plt.xlabel("Clase (0 = legítima, 1 = fraude)")
plt.ylabel("Monto ($)")
plt.ylim(0, 500)  # limitamos el eje Y para ver mejor (hay valores extremos)
plt.show()
```

```python
# Convertir Time (segundos) a horas del "día" simulado
df["Hour"] = (df["Time"] // 3600) % 24

plt.figure(figsize=(10,4))
sns.histplot(data=df[df["Class"]==0], x="Hour", bins=24, color="steelblue", label="Legítima", stat="density", alpha=0.5)
sns.histplot(data=df[df["Class"]==1], x="Hour", bins=24, color="red", label="Fraude", stat="density", alpha=0.5)
plt.title("Distribución de transacciones por hora")
plt.xlabel("Hora del día")
plt.ylabel("Densidad")
plt.legend()
plt.show()
```
